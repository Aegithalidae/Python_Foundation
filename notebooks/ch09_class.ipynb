```python
# car.py
class Car():
    '''一次模拟汽车的简单尝试'''
    def __init__(self, make, model, year):
        self.make =make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self):
        '''返回整洁的描述信息'''
        long_name = f"{str(self.year)} {self.make} {self.model}"
        return long_name.title()

    def read_odometer(self): 
        """打印一条指出汽车里程的消息""" 
        print("This car has " + str(self.odometer_reading) + " miles on it.")

    def update_odometer(self, milage):
        '''
        将里程表读数设置为指定的值
        禁止将里程表读数往回调
        '''
        if milage >= self.odometer_reading:
            self.odometer_reading = milage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        '''将里程表读数增加指定的值'''
        self.odometer_reading += miles

my_used_car = Car('subaru', 'outback', 2013)
print(my_used_car.get_descriptive_name())

my_used_car.update_odometer(23500)
my_used_car.read_odometer()

my_used_car.increment_odometer(150)
my_used_car.read_odometer()
```

```python
class User:
    """表示一个普通用户"""

    def __init__(self, first_name, last_name):
        # 用户的基础属性
        self.first_name = first_name
        self.last_name = last_name
        self.full_name = f"{first_name} {last_name}"

    def describe_user(self):
        """输出用户的基本信息"""
        print(f"Hello, {self.first_name.title()}.")


class Privileges:
    """表示一组管理员权限"""

    def __init__(self):
        # 权限列表属于这个 Privileges 对象自身的状态
        self.privileges = [
            'can add post',
            'can delete post',
            'can ban user'
        ]

    def show_privileges(self):
        """显示当前拥有的所有权限"""
        print("User has the following privileges:")
        for privilege in self.privileges:
            print(f"- {privilege}")

    def add_privilege(self, privilege):
        """向权限列表中新增一条权限"""
        self.privileges.append(privilege)
        print(f'The privilege "{privilege}" has been added.')


class Devices:
    """表示一组设备"""

    def __init__(self):
        # 设备列表属于这个 Devices 对象自身的状态
        self.devices = ['keyboard', 'computer']

    def show_devices(self):
        """显示当前拥有的所有设备"""
        print("User has the following devices:")
        for device in self.devices:
            print(f"- {device}")

    def add_device(self, device):
        """向设备列表中新增一台设备"""
        self.devices.append(device)
        print(f'The device "{device}" has been added.')


class Admin(User):
    """管理员是一种特殊的用户"""

    def __init__(self, first_name, last_name):
        # 先初始化父类 User 的属性
        super().__init__(first_name, last_name)

        # 再添加管理员特有的属性
        # 注意：这里的属性不是简单值，而是两个独立对象
        self.privileges = Privileges()
        self.devices = Devices()


# 创建一个管理员对象
admin = Admin('wu', 'kaiyuan')

# 调用“设备对象”中的方法
admin.devices.show_devices()

# 调用“权限对象”中的方法
admin.privileges.show_privileges()

# 为设备列表添加新设备
admin.devices.add_device('iPhone')
admin.devices.add_device('MacBook')

# 再次显示更新后的设备列表
admin.devices.show_devices()

# 为权限列表添加新权限
admin.privileges.add_privilege('can edit post')

# 再次显示更新后的权限列表
admin.privileges.show_privileges()
```
