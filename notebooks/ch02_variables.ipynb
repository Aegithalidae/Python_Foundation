## 2.2.1 变量名的命名和使用

变量名只能包含字母、数字和下划线。变量名可以字母或下划线打头，但不能以数字打头，例如，可将变量命名为message_1，但不能将其命名为1_message。

变量名不能包含空格，但可使用下划线来分隔其中的单词。例如，变量名greeting_message可行，但变量名greeting message会引发错误。 

不要将Python关键字和函数名用作变量名，即不要使用Python保留用于特殊用途的单词，如print

变量名应既简短又具有描述性。例如，name比n好，student_name比s_n好，name_length比length_of_persons_name好。

慎用小写字母l和大写字母O，因为它们可能被人错看成数字1和0。

## 2.3 字符串

字符串就是一系列字符。在Python中，用引号括起的都是字符串，其中的引号可以是单引号，也可以是双引号。

## 2.3.1 使用方法修改字符串的大小写

```python
name = "ada lovelace"
name.title()
print(name.title())
```

```python
name = "ada -loveLace"
print(name.title())
```

```python
print(name.upper())
print(name.lower())
```

## 2.3.2 拼接字符串

将字符串储存在不同的变量中，要显示时再将它们合并

```python
first_name = "ada"
last_name = "lovelace"
print(first_name + " " + last_name)
print("Hello, " + first_name.title() + " " + last_name.title() + "!")
```

```python
name = "Alice"
age = 20

text = f"Name: {name}, Age: {age}"
print(text)
```

## 2.3.3 使用制表符或换行符来添加空白

```python
sentence = "How are you today?"
# print("I said, "\n + sentence)
# 这会报错，\n必须包裹在字符串引号内，或者作为一个独立的字符串连接
print("I said, \n" + sentence)
print("\tI said, \n" + sentence)
# \n表示换行，\t表示制表符（tab），可以用来增加缩进
```

## 2.3.4 删除空白

```python
favorite_language = '  python   '
print(favorite_language)
print(favorite_language.rstrip())
print(favorite_language.lstrip())
print(favorite_language.strip())
print(favorite_language) # 上面3个方法并没有改变原字符串，除非你将结果赋值回变量
favorite_language = favorite_language.strip() # 现在变量已经被更新了
print(favorite_language)
```
