# 6.2 使用字典

```python
# 字典是一系列键值对，即每一个键对应一个值
# 键是唯一的，值可重复且可以是任何数据类型
# 字典使用花括号 {} 定义，键值对之间用冒号 : 分隔，多个键值对之间用逗号 , 分隔
# 字典是无序的，不能通过索引访问，但可以通过键来访问对应的值
# 字典是可变的，可以添加、修改和删除键值对
# 字典的键必须是不可变类型，如字符串、数字或元组，而值可以是任何类型
alien_0 = {'color': "green", 'points': 5}
print(alien_0['color'])
print(alien_0['points'])
```

```python
# 添加键值对
person = {}
person['name'] = str('alice'.title())
person['age'] = 30
person['city'] = 'New York'
other = {'height' : 165, 'weight' : 60}
person.update(other)
print(person)
```

```python
# 追踪外星人
alien_0 = {'x_position': 0, 'y_position': 0, 'speed': 'medium', 'color': 'green'}
print(alien_0)
print(f"Original position: ({alien_0['x_position']}, {alien_0['y_position']})")

# 检查速度并根据快慢提供坐标增量
if alien_0['speed'] == 'slow':
    x_increment = 5
    y_increment = 5
elif alien_0['speed'] == 'medium':
    x_increment = 10
    y_increment = 10
elif alien_0['speed'] == 'fast':
    x_increment = 15
    y_increment = 15
else :
    x_increment = 0
    y_increment = 0
    print("Warning: Unknown speed. Alien remains stationary.")

# 将坐标增量添加到原始坐标
alien_0['x_position'] += x_increment
alien_0['y_position'] += y_increment
print(f"\n{alien_0}")
print(f"Now position: ({alien_0['x_position']}, {alien_0['y_position']})")
```

```python
# 追踪外星人
alien_0 = {'x_position': 0, 'y_position': 0, 'speed': 'medium', 'color': 'green'}
speed_map = {'low': 5, 'medium': 10, 'fast': 15}
print(alien_0)
print(f"Original position: ({alien_0['x_position']}, {alien_0['y_position']})")

# 使用 .get() 方法，如果找不到对应的速度，默认增量为 0
x_increment = speed_map.get(alien_0['speed'], 0)
y_increment = speed_map.get(alien_0['speed'], 0)
alien_0['x_position'] += x_increment
alien_0['y_position'] += y_increment

print(f"\n{alien_0}")
print(f"Now position: ({alien_0['x_position']}, {alien_0['y_position']})")
```

```python
# 6-6 investigation
favorite_languages = { 
    'Jen': 'python', 
    'saRaH': 'c', 
    'edward': 'ruby', 
    'phil': 'python', 
    } 

normalize_dict = {k.lower(): v for k, v in favorite_languages.items()}
operator_name = ['jEn', 'mIke', 'sarah', 'john']

for name in operator_name:
    if name.lower() in normalize_dict:
        print(f"Thank you {name.lower().title()}, we appreciate your contribution")
    else:
        print(f"Please accept the investigation, {name.lower().title()}")
```

```python
# 6-6 调查：处理不规范的字典键名与名单匹配

# 原始调查数据：模拟现实中大小写不规范的录入情况
favorite_languages = { 
    'Jen': 'python', 
    'saRaH': 'c', 
    'edward': 'ruby', 
    'phil': 'python', 
} 

# 数据标准化（Normalization）：
# 使用字典推导式创建一个键全部为小写的新字典。
# 这样做是为了在后续查询中实现 O(1) 的常数级查找效率，同时解决大小写匹配问题。
normalize_dict = {k.lower(): v for k, v in favorite_languages.items()}

# 待检查的参与者名单（包含不同的大小写格式）
operator_name = ['jEn', 'mIke', 'sarah', 'john']

for name in operator_name:
    # 统一转为小写，以便与标准化后的字典键进行比对
    clean_name = name.lower()
    
    # 利用字典（哈希表）的高性能查找确认该人是否已完成调查
    if clean_name in normalize_dict:
        # title() 让输出的名字格式更统一美观（首字母大写）
        print(f"Thank you {clean_name.title()}, we appreciate your contribution")
    else:
        print(f"Please accept the investigation, {clean_name.title()}")
```

```python
# 6-6 investigation
favorite_languages = { 
    'jen': 'python', 
    'sarah': 'c', 
    'edward': 'ruby', 
    'phil': 'python', 
    } 

check_names = [i.lower() for i in favorite_languages]
operator_name = ['jEn', 'mIke', 'sarah', 'john']
# 字典本就为查询而生，将其键转换为列表有些多余了，复杂度变成了O(n)

for name in operator_name:
    if name.lower() in check_names:
        print(f"Thank you {name.lower().title()}, we appreciate your contribution")
    else:
        print(f"Please accept the investigation, {name.lower().title()}")
```

# 6.4 嵌套

```python
alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
    print(alien)
```

```python
aliens = []

for i in range(1, 11):
    new_alien_green = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien_green)

for i in range(11, 21):
    new_alien_yellow = {'color': 'yellow', 'points': 10, 'speed': 'medium'}
    aliens.append(new_alien_yellow)

for i in range(21, 31):
    new_alien_red = {'color': 'red', 'points': 15, 'speed': 'fast'}
    aliens.append(new_alien_red)

for i in aliens[:]:
    print(i)
```

```python
users = { 
    'aeinstein': { 
        'first': 'albert', 
        'last': 'einstein', 
        'location': 'princeton', 
        }, 
    'mcurie': { 
        'first': 'marie', 
        'last': 'curie', 
        'location': 'paris', 
        }, 
    }

for usernames, user_info in users.items():
    print(f"Username is : {usernames}")
    full_name = f"{user_info['first'].title()} {user_info['last'].title()}"
    location = user_info['location']

    print(f"Full name : {full_name}")
    print(f"Location : {location.upper()}")
    print('\n')
```
