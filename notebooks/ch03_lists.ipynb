# 3.1 列表是什么

```python
# bicycles.py
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

列表用方括号`[]`表示，其中可以有多个元素，元素之间用`,`区分。有点像集合

**注意**：列表中的元素是有**顺序**的，专业一点叫**索引**
下面注意看索引从0起始而非1

```python
print(bicycles[0])
print(bicycles[1])
print(bicycles[2])
print(bicycles[3])
```

通过索引，可以直接使用列表中的元素

```python
print(f"My first bicycle was a {bicycles[0].title()}.")
```

# 3.2 修改列表元素

```python
motorcycles = ['honda', 'yamaha', 'suzuki'] 
print(motorcycles)

# 直接给列表元素赋新值
motorcycles[0] = 'ducati'
print(motorcycles)
# 可以用这种方法修改任意列表元素
motorcycles[1] = 'harley-davidson'
motorcycles[2] = 'bmw'
print(motorcycles)
```

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
motorcycles.append('ducati')
print(motorcycles)
```

# 3.2.2 在列表中添加元素

```python
# 可以用append()方法在列表末尾添加元素，()里是要添加的元素
motorcycles.append('ducati')
print(motorcycles)

# 可以用这个方法给空列表添加元素
# 这种方法很有用，因为你可能不知道要添加哪些元素，或者要添加多少元素
```

# 3.2.3 在列表中插入元素

```python
# 使用方法insert()在列表中插入元素，()里是要插入的元素的位置和要插入的元素
# 是将元素插入位置，而不是替换位置上的元素
motorcycles.insert(0, 'ducati')
print(motorcycles)
```

# 3.2.4 删除列表中的元素

```python
# 如果知道元素的索引但不知道元素的值，可以使用del语句删除元素
del motorcycles[0]
print(motorcycles)

# 再次删除元素
del motorcycles[0]
print(motorcycles)

# 注意，del语句永久删除列表中的元素，无法恢复
# 所以下面的运行结果会不同，因为之前的代码已经删除了元素
```

```python
motorcycles = ['honda', 'yamaha', 'suzuki'] 
print(motorcycles) 

# 有时候，删除的元素可能还需要使用，可以使用方法pop()删除元素
# pop()方法默认删除列表末尾的元素，并返回这个元素的值
popped_motorcycle = motorcycles.pop()
print(motorcycles)
print(popped_motorcycle)
# 也可以指定要删除的元素的索引,在()里指定要删除的元素的索引即可
first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}.")
print(motorcycles)
# pop()方法删除的元素仍然可以使用，因为它返回了这个元素的值
```

```python
motorcycles = ['honda', 'yamaha', 'suzuki'] 
print(motorcycles)

# 如果不知道要删除哪个元素，但知道要删除的元素的值，可以使用方法remove()
# 在()里指定要删除的元素的值即可

motorcycles.remove('yamaha')
print(motorcycles)
# remove()方法只删除列表中第一个指定的值
# 如果要删除的值在列表中出现了多次，必须使用循环来删除所有指定的值
# 注意，方法remove()在删除元素时不会返回这个元素的值
removed_motorcycle = motorcycles.remove('suzuki')
print(removed_motorcycle)
# 运行结果是None，因为方法remove()没有返回值
# 但不会报错
```

# 3.3 组织列表

列表内元素有“顺序”这一概念，使用恰当的方法可以有序地安排列表元素

```python
# cars.py
cars = ['bmw', 'audi', 'toyota', 'subaru']
print(cars)
cars.sort()
print(cars)

# 方法sort()永久修改列表元素的顺序，并且没有返回值
# 下面的运行结果是None，因为方法sort()没有返回值
sorted_cars = cars.sort()
print(sorted_cars)

# 方法sort()默认的顺序是按字母顺序
# 可以传递参数reverse=True来按相反的字母顺序排序
cars.sort(reverse = True)
print(cars)
print(cars.sort())
print(cars)
```

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
# 函数sorted()临时排序列表元素，并返回这个列表的副本
print("Here is the original list:")
print(cars)
print("\nHere is the sorted list:")
print(sorted(cars))
print("\nHere is the original list again:")
print(cars)
```

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
# 可以倒着打印列表元素，有专门的方法.reverse()，这个方法永久修改列表元素的顺序
print("Here is the original list:")
print(cars)
cars.reverse()
print("\nHere is the reversed list:")
print(cars)

# 补充一个确定列表长度的函数len()
length = len(cars)
print(length)
# len()返回的是整数而非字符串，所以不能直接在字符串中使用这个函数
print(type(length))
```

```python
# 3-8 seeing the world
trip_locations = ['paris', 'new york', 'tokyo', 'sydney', 'rome']
print(trip_locations)
print(sorted(trip_locations))
print(trip_locations)
print(sorted(trip_locations, reverse = True))
print(trip_locations)
# reverse_locations = trip_locations.reverse()
# 运行结果是None，因为方法reverse()没有返回值
trip_locations.reverse()
print(trip_locations)
trip_locations.reverse()
print(trip_locations)
trip_locations.sort()
print(trip_locations)
trip_locations.sort(reverse = True)
print(trip_locations)
```
