# 4.1 遍历列表

```python
# magicians.py
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(magician)
```

直白易懂。想要逐个操作列表中的元素，使用for循环是个不错的主意。
---
这个 `for` 循环之所以能自动从头遍历到结尾，本质上不是它自己在“数索引”，而是它在背后调用了一个迭代器机制：当执行 `for magician in magicians` 时，Python 会先把列表转换成一个迭代器对象，这个对象内部维护当前位置并从第一个元素开始；随后每一轮循环都会通过 `next()` 取出下一个元素，直到元素取完为止，此时迭代器会抛出 StopIteration 异常，`for` 捕获这个信号后自然结束循环，因此“从哪里开始”和“在哪里停止”其实都由迭代器的内部状态和耗尽机制决定，而不是 `for` 本身在做计数控制。

# 4.3.2 `range()`创建数字列表

```python
# list()函数可以将range对象转换为列表
numbers = list(range(1, 6))
print(numbers)
# 偶数列
even_numbers = list(range(2, 11, 2))
print(even_numbers)
# 奇数列
odd_numbers = list(range(1, 10, 2))
print(odd_numbers)
```

```python
# `range()`函数也可以生成一个指定范围内的数字的平方列表
squares = []
# range()生成的数字可以直接用于计算，不需要先转换为列表
for value in range(1, 11):
# for循环实际上是在迭代range对象生成的数字序列，每次迭代都会将当前数字赋值给变量value
    squares.append(value ** 2)
    # 括号里不仅可以是简单表达式，还可以是函数调用，甚至是更复杂的组合表达式
    # 括号里要能返回一个值，即表达式必须是一个合法的Python表达式，不能是语句
print(squares)
```

# 4.4 切片

```python
# 可以选择部分使用列表元素，比如打印前几个元素，或者最后几个元素，或者列表的中间部分
# 这种操作就叫作切片（slice）
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[0:3])  # 输出前三个元素，索引规则是左闭右开
# 若不指定第一个索引，默认从0开始
print(players[:4])  # 输出前四个元素
# 若不指定第二个索引，默认一直到列表末尾
print(players[2:])  # 输出从索引2开始到末尾的所有
# 负数索引表示从列表末尾开始计数，-1表示最后一个元素，-2表示倒数第二个，以此类推
print(players[-3:])  # 输出最后三个元素
```

```python
# 切片也是列表
# 遍历切片就是遍历一个列表
players = ['charles', 'martina', 'michael', 'florence', 'eli']

print("Here are the first three players on my team:") 
for player in players[:3]:
    print(player.title())
```

```python
seq = [i for i in range(1, 11)]
print(seq[::2])
```

```python
# 切片还可以用来复制列表
# 创建一个[:]切片会返回一个包含列表中所有元素的新列表
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]
print("My favorite foods are:")
print(my_foods)
print("\nMy friend's favorite foods are:")
print(friend_foods)
my_foods.append('cannoli')
friend_foods.append('ice cream')
print('\n')
print(my_foods)
print(friend_foods)
# 注意：如果直接将一个列表赋值给另一个变量，比如friend_foods = my_foods
# 那么两个变量将指向同一个列表对象，对其中一个变量的修改会影响另一个变量
# 因为它们引用的是同一个列表对象。而使用切片创建新列表则可以避免这个问题
# 在使用Python时，一定要区分变量名和对象本体
```

```python
# 4-10 slices
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print("The first three items in the list are:")
print(players[:3])
print("\nThree items from the middle of the list are:")
print(players[1:4])
print("\nThe last three items in the list are:")
print(players[-3:])
```

# 4.5 元组

```python
# 元组内的值不可修改，可用作常量
# 元组用圆括号表示，列表用方括号表示
dimensions = (200, 50)
print(dimensions[0])  # 输出第一个元素
print(dimensions[1])  # 输出第二个元素
print(dimensions[-1])  # 输出最后一个元素
# dimensions[0] = 250  # 这行代码会报错，因为元组内的值不可修改
```

```python
# 显然元组是一个可迭代对象，可以使用for循环来遍历元组中的元素
dimensions = (200, 50)
for dimension in dimensions:
    print(dimension)
```

```python
# 虽然不能修改元组中的元素，但可以给变量重新赋值来改变整个元组
dimensions = (200, 50)
print("Original dimensions:")
for dimension in dimensions:
    print(dimension)
dimensions = (400, 100)
print("\nModified dimensions:")
for dimension in dimensions:
    print(dimension)
# 元组(200, 50)已经成为不可达对象，等待垃圾回收机制回收
# 所以在编程时，要区分变量名和对象本体，理解变量名只是一个标签，可以指向不同的对象，而对象本体才是实际的数据结构
# 如果途中遗失了对某个对象的引用，那么这个对象就会成为不可达对象，等待垃圾回收机制回收
```

```python
t = ([1, 2], [3, 4])
t[0][0] = 999
print(t)
```
