# 5.2 条件测试

```python
# if语句的核心是条件测试，条件测试的结果总是True或False。
# 根据条件测试的结果，程序决定是否执行if语句中的代码块。
# 也就是说，if语句结构包括布尔值判断和代码执行两个部分
# 判断两个值相等，用==;判断两个值不相等，用!=
# 判断一个值是否包含在列表中，用in;
# 判断一个值是否不包含在列表中，用not in
requested_toppings = ["mushrooms"]

if requested_toppings == "mushrooms":
    print("Hold the anchovies!")
print(requested_toppings)

# 还有其他常用判断比较
# 判断数值比较：<、>、<=、>=
# 判断多个条件：and、or
# 判断是否相等：==、!=
# 判断是否包含：in、not in
```

```python
# and是与，or是或
# and要求所有条件都满足，or要求至少一个条件满足
print(5 > 3 and 2 < 4)  # True
print(5 > 3 and 2 > 4)  # False
print(5 > 3 or 2 >4)    # True
print(5 < 3 or 2 >4)    # False
# python中不用0和1来表示布尔值，直接使用True和False
```

```python
# 存在与不存在的判断
banned_users = ["andrew", "carolina", "david"]
user = "marie"

if user not in banned_users:
    print(f"{user.title()}, you are banned!")
else:
    print(f"{user.title()}, you can post a response if you wish.")
```

# 5.3 if语句

```python
# if语句的结构是先进行条件测试，然后根据测试结果执行相应的代码块。
# 条件测试的结果是布尔值True或False，根据这个结果，程序决定是否执行if语句中的代码块。
# if conditional_test:
#     do something
# 仅当条件测试的结果为True时，才执行if语句中的代码块，否则跳过该代码块继续执行后续代码。
```

```python
age = 19
if age >=18:
    print("You are old enough to vote!")
else:
    print("Sorry, you are too young to vote!")

# 这是一个简单的if-else结构，根据年龄判断是否有投票资格。
# if-else结构允许程序在条件测试为False时执行另一段代码块，提供了更完整的逻辑处理。
```

```python
# 当条件不是简单的是与否时，可以使用elif检查多个条件
# 注意：if-elif-else的结构是互斥的，一次运行只会执行其中一个条件的代码块，其余将被跳过。
# 下面是一个公园按年龄收费的例子
age = 12

if age < 4:
    print("You are free to enter the park!")
elif age > 4 and age <18:
    print("You need to pay $5 to enter the park.")
else:
    print("You need to pay $10 to enter the park.")
```

```python
# elif是else if的缩写，可以根据需要使用任意数量的elif
age = 12 

if age < 4:
    price = 0
elif age < 18:
    price = 5
elif age <65:
    price = 10
else:
    price = 5
print(f"You need to pay ${price} to enter the park.")
```

# 5.4 使用if处理列表

```python
requested_toppings = ['mushrooms', 'green peppers', 'extra cheese']

for requested_topping in requested_toppings: 
    if requested_topping == 'green peppers': 
        print("Sorry, we are out of green peppers right now.") 
    else:
        print("Adding " + requested_topping + ".")  

print("\nFinished making your pizza!") 
```

```python
available_toppings = ['mushrooms', 'olives', 'green peppers', 
                    'pepperoni', 'pineapple', 'extra cheese'] 
requested_toppings = ['mushrooms', 'french fries', 'extra cheese']
checked_toppings = []

for requested_topping in requested_toppings:
    if requested_topping in available_toppings:
        print(f"Adding {requested_topping}.")
        checked_toppings.append(requested_topping)
    else:
        print(f"Sorry, we don't have {requested_topping}.")

print(f'''\nHere are your toppings:{checked_toppings}.
Please wait while we are preparing your pizza!''')
```

```python
# 5-9 no users
users = ['admin', 'alice', 'bob', 'charlie', 'david']

for user in users:
    del users[0]
    if users:
        print("We have some users!")
    else:
        print("We need to find some users!")
print(users)
# 注意，在循环中删除列表元素可能会导致意外行为，因为列表的长度在迭代过程中发生变化。
# 在这个例子中，循环会删除列表中的元素，但由于列表长度在迭代过程中发生变化，可能会导致某些元素被跳过或引发错误。
# 更安全的做法是在循环外部检查列表是否为空，而不是在循环内部删除元素。
# 这个例子中，第一次循环时，会删除'admin'并打印"We have some users!"
# 但注意，第二次循环开始时，迭代器索引从0指向1，此时列表的索引1并不是原来的'alice'，而是'bob'
# 因此，'alice'会被跳过，导致输出结果不正确。
# 同理，第三次循环时，索引指向2，此时列表的索引2是'david'，而不是'charlie'，导致'charlie'也被跳过。
# 最后一次循环时，索引指向3，此时列表的索引3已经不存在，会退出遍历。
# 因此，users的元素还剩下['charlie', 'david']，而不是被完全清空。
```

```python
# 5-9 no users
users = ['admin', 'alice', 'bob', 'charlie', 'david']

while users:
    users.pop()
    if users:
        print("We have some users!")
    else:
        print("We need to find some users!")
print(users)
# 这个例子中，使用while循环和pop方法来逐个删除列表中的元素，直到列表为空。
# 每次循环时，pop方法会删除列表中的最后一个元素，并返回该元素的值。
# 当列表不为空时，打印"We have some users!"；当列表为空时，打印"We need to find some users!"。
# 最后，打印空列表[]，表示所有用户都已被删除。
# 相比遍历同时删除元素的方式，这种方法更安全，不会导致跳过元素或引发错误。
```

```python
# 5-10 checking usernames
current_users = ['admin', 'alice', 'bob', 'charlie', 'david']
current_users_lower = {user.lower() for user in current_users}
# 遇事不决先转换成小写再比较，可以避免大小写导致的用户名冲突问题。
# 新建立一个检查列表是比较好的想法
# 用集合比列表更高效，因为集合的查找时间复杂度为O(1)，而列表的查找时间复杂度为O(n)。
new_users = ['eve', 'frank', 'alice', 'grace', 'bob']

for new_user in new_users:
    if new_user.lower() in current_users_lower:
        print(f"Sorry, the username '{new_user}' is already taken. Please choose a different username.")
    else:
        print(f"The username '{new_user}' is available.")
```

```python
# 5-11 ordinal numbers
numbers = list(range(1, 10))
for number in numbers:
    if number == 1:
        print(f"{number}st")
    elif number == 2:
        print(f"{number}nd")
    elif number == 3:
        print(f"{number}rd")
    else:
        print(f"{number}th")
```
